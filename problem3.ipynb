```python
import random
import time

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:

        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            return -1, probes

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < 0 or pos >= len(arr):
            return -1, probes

        if arr[pos] == key:
            return pos, probes

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


def manual_input():
    n = int(input("Enter number of elements: "))
    arr = []

    print("Enter elements:")
    for _ in range(n):
        arr.append(int(input()))

    arr.sort()
    return arr


def generate_uniform(n):
    start = random.randint(1, 50)
    gap = random.randint(2, 10)
    return [start + i * gap for i in range(n)]


def generate_non_uniform(n):
    arr = []
    value = 1

    for i in range(n):
        value += random.randint(1, i + 5)
        arr.append(value)

    return arr


print("\nInterpolation Search Analysis")
print("1. Uniform Dataset")
print("2. Non-Uniform Dataset")

choice = int(input("Choose dataset type: "))

print("\n1. Manual Input")
print("2. Random Generation")

method = int(input("Choose input method: "))

if method == 1:
    arr = manual_input()
else:
    n = int(input("Enter dataset size: "))

    if choice == 1:
        arr = generate_uniform(n)
    else:
        arr = generate_non_uniform(n)

print("\nDataset:")
print(arr)

key = int(input("\nEnter search key: "))

start_time = time.perf_counter()

index, probes = interpolation_search(arr, key)

end_time = time.perf_counter()

execution_time = (end_time - start_time) * 1000000

print("\n----- Result -----")

if index != -1:
    print("Key Found")
    print("Index Position:", index)
else:
    print("Key Not Found")

print("Number of Probes/Comparisons:", probes)
print("Execution Time: {:.3f} microseconds".format(execution_time))
```
